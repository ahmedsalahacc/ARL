```python
import numpy as np
import pandas as pd
```

## Exploring the Dataset

```python
df = pd.read_csv('train.csv')
df.head(20)
```

```python
df.info()
```

```python
### Displaying the range of values of each string column
for col,ttype in zip(df.columns, df.dtypes):
    if ttype=="object":
        print(f'{col}:\t{set(df[col].values)}')
```

###  Comparing the effect of null values on the dataframe

```python
print(df.shape)
print(df.isna().sum())
```

I will remove all columns that have more than 100 null values and will replace all the null  values that have string values with the mode of the column, and the nulls in the integer columns will be replaced by the mean of the values

```python
for col, count, dtype in zip(df.columns,df.isna().sum().values, df.dtypes):
    if count>=1 and count<100:
        if dtype =='object':
            df[col] = df[col].astype('|S') # change datatype to string
            df[col].fillna(df[col].mode(), inplace=True)
        else:
            df[col].fillna(df[col].mean(), inplace=True)
    elif count>=100:
        df.drop(columns = [col], inplace=True)
```

```python
# sanity check
df.isna().sum().sum()
```

### Onehot encoding

```python
# one hot encoding the data
for col, dtype in zip(df.columns,df.dtypes):
    if dtype=='object':
        one_hot = pd.get_dummies(df[col])
        # dropping the col
        df.drop(col,axis=1,inplace=True)
        # joining the new value
        df = df.join(one_hot,how='left',lsuffix='_left')
```

```python
# sanity check
df.head()
```

# Preprocessing

```python
# get the index of Y
out_index = df.columns.tolist().index('SalePrice')
y = df.iloc[:,out_index].values.reshape(-1,1)
df.drop(columns=['SalePrice'], inplace=True)
X = df.iloc[:,:].values
print(X.shape)
print(y.shape)
```

```python
# adding the extra column for ones (bias)
X = np.hstack((np.ones(y.shape),X))
# sanity check
X.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25)
```

```python
print("Train X shape:", X_train.shape)
print("Train y shape:", y_train.shape)
print("Test X shape:", X_test.shape)
print("Test y shape:", y_test.shape)
```

# Linear Regression

## My Model

```python
class LinearRegression:
    def __init__(self):
        self.theta = None
    
    def train(self, X, y):
        self.theta = np.linalg.pinv(X)@y
        return self.theta
    
    def predict(self, X):
        return X@self.theta
```

```python
reg = LinearRegression()
reg.train(X_train,y_train)
y_pred = reg.predict(X_test)
```

## Scikit learn

```python
from sklearn.linear_model import LinearRegression
sklearn_reg = LinearRegression()
sklearn_reg.fit(X_train,y_train)
y_pred2 = sklearn_reg.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
print("My model ASE")
print(mean_absolute_error(y_test,y_pred))
print("Sklearn model ASE")
print(mean_absolute_error(y_test,y_pred2))
```
